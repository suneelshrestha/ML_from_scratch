# linear regression with single variable
formula
y = mx + b


```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = {
    "area":[8,10,12,16,18],
    "price":[200,300,400,500,550]
}
df = pd.DataFrame(data)
df
```

```python
plt.title("linear regression")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price)
```

```python
#use the linear regression
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
#predict the price for the area 23
reg.predict([[23]])
```

```python
#how it works
# y = mx + b
# where m is coefficient of gradient and b = y-intercept_
m = reg.coef_
b = reg.intercept_

#now to predict value be x
x = 23

#then by formula
y = m*x + b
print(y)
```

```python
plt.title("linear regression")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price)
plt.plot(df.area,reg.predict(df[['area']]),color="red")
```
