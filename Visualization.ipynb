# Visualization using Matplotlib and seaborn

## Matplotlib

```python
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#create some data to plot
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

#plot the values
plt.plot(x,y)
plt.show()
```

```python
#ploting multiple data and also change the color or line and the shape that represent by default it is represented by line
#this r will show in red and + sign will be used
w = np.linspace(-10,10,100)
plt.plot(x,y,'r+')
plt.plot(x,z,'b.')
plt.plot(x,np.sin(w),'g*')
plt.show()
```

```python
#ploting parabola
y = np.linspace(-10,10,20)
z = y**2
plt.plot(y,z,'r+')
plt.show()
```

```python
#creating a df and plotting them
data ={
    "cars":["byd","tesla","farari"],
    "stock":[10,5,8]
}

#convert the dict to pandas dataframe
df = pd.DataFrame(data)
```

```python
#not plot the data in bar graph with lables and the title
plt.bar(df['cars'],df['stock'])
plt.ylabel("No of stocks of the cars")
plt.xlabel("Brand of the cars")
plt.title("Bar graph of cars and their availabe stocks")
plt.show()
```

```python
#showing the df in pitchart
# big size for better
fig = plt.figure(figsize=(7,7))
plt.pie(df['stock'], labels=df['cars'], autopct='%1.1f%%')
plt.ylabel("No of stocks of the cars")
plt.xlabel("Brand of the cars")
plt.title("pie chart of cars and their availabe stocks")
plt.show()
```

## seaborn


```python
#import seaborn
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#print the available datasets
print(sns.get_dataset_names())
```

```python
#load one datasets
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',col='time', hue='smoker', style='smoker', size='size')
```

```python
#load iris dataset
iris = sns.load_dataset('iris')
iris
```

```python
#plot iris data
sns.relplot(data=iris, x='sepal_length',y='petal_length',hue='species',size='sepal_width')
```

```python
#other dataset
# sns.displot(data=iris,x='sepal_width' y='sepal_length',col='sepal_width')
sns.histplot(data=iris,x='sepal_length',y='petal_length',hue='sepal_width')
```

```python
#get the cmap list
sns.lineplot(data=iris,x='sepal_length',y='petal_length')
plt.show()
```

```python
#dislpot
sns.distplot(iris['petal_length'])
```

```python
#scatter
sns.scatterplot(data=iris , x='sepal_length', y='petal_length', hue='sepal_width')
```
