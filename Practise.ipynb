```python
#display titles 0f moview with runtime >= 100
new_df=df[df['Runtime (Minutes)']>=100]
new_df['Title']
new_df
```

```python
#find the year with the highest average voting
avg_df = df.groupby('Year')['Votes'].mean()
avg_df
```

```python
#find the year with the highest average Revenue (Millions)
avg_rev_df = df.groupby('Year')['Revenue (Millions)'].mean()
avg_rev_df
```

```python
#find the year rating for each director
year_df=df.groupby('Director')['Rating'].mean()
year_df
```

```python
sorted_df=df.sort_values('Runtime (Minutes)', ascending=False)
sorted_df[['Title','Runtime (Minutes)']]
```

```python
df.groupby('Year').count()
```

```python
#find the most popular movie title (highest Revenue)
max_rev_df = df.groupby('Title')['Revenue (Millions)'].max()
max_rev_df.sort_values(ascending=False)
```

```python
#display the highest moview rated with theri title and theri rating
sortedVal = df.groupby('Title')['Rating'].max()
sortedVal.sort_values(ascending=False)
```

```python
#find the average revenue by genre
avg_rev_df =df.groupby(['Title','Genre'])['Revenue (Millions)'].median()
avg_rev_df.sort_values(ascending=False)
```

```python
#find the unique no of genre in the dataset
unique_genres_list = df['Genre'].unique()
print(unique_genres_list)
```

```python
#count the no of movies released in each genre
each_gen_mov= df.groupby('Genre')['Title'].count()
each_gen_mov
```

```python
#find the top 10 director with the most movies
d_with_most_movies = df.groupby('Director')['Title'].count()
d_with_most_movies.sort_values(ascending=False).head(10)
```

```python
#find the avg rating in each year
avg_rat = df.groupby('Year')['Rating'].median()
avg_rat
```

```python
#find the total revenu generated each year
rev=df.groupby('Year')['Revenue (Millions)'].sum()
rev.sort_values(ascending=False)
```

```python
#Find the Most Common Actor in the Dataset
actor = df['Actors'].value_counts().idxmax()
actor
```

```python
#Find the Average Runtime of Movies in Each Genre
avg=df.groupby('Genre')['Runtime (Minutes)'].median()
avg.sort_values(ascending=False)
```
