```python
#read data
import pandas as pd

df = pd.read_csv("/content/sample_data/world_population.csv")
df
```

## filtering data
single condition and multi conditon


```python
#filtered rows = df[df["salary"]> 500]  //single condtion
#filtered_rows = df[(df["salary"] > 500) & (df["age"] > 30)]

#this is for any condtion between two
#filtered_rows = df[(df["salary"] > 500) | (df["age"] > 30)]
```

```python
#display the list of counties having rank <= 10
df=df[df['Rank']<= 10]
df[['Rank','Country']]
```

```python
#Create a list of specific country ['Bangladesh', 'Brazil'] and return df which falls under the specific list only
# df=df[(df['Country'] == 'Brazil') | (df['Country'] == 'Bangladesh')]
# df

#or
county =['Bangladesh','Brazil']
df = df[df["Country"].isin(county)]
df
```

```python
# apply filter which has United in its name
df = df[df['Country'].str.contains('United')]
df
```

```python
#Create a dataframe df2 and give Country as its index
df2 = df.copy()

# df2 = df2.set_index('Country')
# df2

#### OR
df2 = df2.set_index(df.Country)
df2
```

```python
#Apply filter Continent and CCA3 in df2 (need only two columns)
df2[['Continent','CCA3']]
```

```python
#Apply filter to see details of Zimbabwe entire row
print(df2.loc['Zimbabwe'])
```
