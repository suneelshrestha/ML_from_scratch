```python
import numpy as np
import pandas as pd
```

```python
#read the data
df = pd.read_excel("/content/sample_data/Customer Call List.xlsx")
df
```

```python
#remove unewanted sign form the lastname
df['Last_Name']=df['Last_Name'].str.replace("...","").str.replace("/","").str.replace("_","").dropna()
df
```

```python
#delete duplicates
df.drop_duplicates(inplace=True)
df
```

```python
#drop column "not useful"
df.drop(columns='Not_Useful_Column', inplace=True)
df
```

```python
#in column paying customer replace yes with y and no with N and replace n/a with blank
df['Paying Customer']=df['Paying Customer'].str.replace("Yes","Y").str.replace("No","N").str.replace("N/a","")
df
```

```python
#replace NaN and na with blank in column Do-not-contact
df['Do_Not_Contact']=df['Do_Not_Contact'].fillna(" ")
df
```

```python
#keep the phonenuber in 3-3-3-remaining digits format
df
```

```python
#change the phone no into 3-3-3-remaining format

#first replace all unwanted symboles from the phonenumber
df['Phone_Number'] = df['Phone_Number'].str.replace("-","").str.replace("/","").str.replace("|","").str.replace("nan","").str.replace("Na","").fillna("")


#now change the format using lambda
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:9] + "-" + x[9:])
df
```

```python
#export the cleaned csv file named as cleanedcalllistdata
df.to_csv("Cleaned_Call_list.csv")
```
